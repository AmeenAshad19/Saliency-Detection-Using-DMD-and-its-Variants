<h1><center>Colour based representation of images</center></h1>

```python
import cv2
import matplotlib.pyplot as plt
```

<h3>Sample Images</h3>

```python
img1 = cv2.cvtColor(cv2.imread('Color Space Analysis Dataset//Dog_blur_bg.jpg'), cv2.COLOR_BGR2RGB)
plt.subplot(1,2,1)
plt.imshow(img1)
plt.title('Blur Background')
plt.axis('off')

img2 = cv2.cvtColor(cv2.imread('Color Space Analysis Dataset//Dog_normal_bg.jpg'), cv2.COLOR_BGR2RGB)
plt.subplot(1,2,2)
plt.imshow(img2)
plt.title('Distinct Background')
plt.axis('off')

plt.show()
```

<h3>Colour Space Analysis - Blur Background</h3>

```python
# Converting RGB Colorspace to YUV Colorspace
img1_YUV = cv2.cvtColor(img1, cv2.COLOR_BGR2YUV)
Y1,U,V = cv2.split(img1_YUV)

# Converting RGB Colorspace to LAB Colorspace
img1_LAB = cv2.cvtColor(img1, cv2.COLOR_BGR2LAB)
L,a,b = cv2.split(img1_LAB)

# Converting RGB Colorspace to LAB Colorspace
img1_YCrCb = cv2.cvtColor(img1, cv2.COLOR_BGR2YCrCb)
Y2, Cr, Cb = cv2.split(img1_YCrCb)

# Representing the image through various color channels
plt.subplot(3,3,1)
plt.imshow(Y1,cmap="gray")
plt.title('Y1 - channel')
plt.axis('off')

plt.subplot(3,3,2)
plt.imshow(U,cmap="gray")
plt.title('U - channel')
plt.axis('off')

plt.subplot(3,3,3)
plt.imshow(V,cmap="gray")
plt.title('V - channel')
plt.axis('off')

plt.subplot(3,3,4)
plt.imshow(L,cmap="gray")
plt.title('L - channel')
plt.axis('off')

plt.subplot(3,3,5)
plt.imshow(a,cmap="gray")
plt.title('a - channel')
plt.axis('off')

plt.subplot(3,3,6)
plt.imshow(b,cmap="gray")
plt.title('b - channel')
plt.axis('off')

plt.subplot(3,3,7)
plt.imshow(Y2,cmap="gray")
plt.title('Y2 - channel')
plt.axis('off')

plt.subplot(3,3,8)
plt.imshow(Cr,cmap="gray")
plt.title('Cr - channel')
plt.axis('off')

plt.subplot(3,3,9)
plt.imshow(Cb,cmap="gray")
plt.title('Cb - channel')
plt.axis('off')

plt.show()
```

<h3>Colour Space Analysis - Distinct Background</h3>

```python
# Converting RGB Colorspace to YUV Colorspace
img2_YUV = cv2.cvtColor(img2, cv2.COLOR_BGR2YUV)
Y1,U,V = cv2.split(img2_YUV)

# Converting RGB Colorspace to LAB Colorspace
img2_LAB = cv2.cvtColor(img2, cv2.COLOR_BGR2LAB)
L,a,b = cv2.split(img2_LAB)

# Converting RGB Colorspace to LAB Colorspace
img2_YCrCb = cv2.cvtColor(img2, cv2.COLOR_BGR2YCrCb)
Y2, Cr, Cb = cv2.split(img2_YCrCb)

# Representing the image through various color channels
plt.subplot(3,3,1)
plt.imshow(Y1,cmap="gray")
plt.title('Y1 - channel')
plt.axis('off')

plt.subplot(3,3,2)
plt.imshow(U,cmap="gray")
plt.title('U - channel')
plt.axis('off')

plt.subplot(3,3,3)
plt.imshow(V,cmap="gray")
plt.title('V - channel')
plt.axis('off')

plt.subplot(3,3,4)
plt.imshow(L,cmap="gray")
plt.title('L - channel')
plt.axis('off')

plt.subplot(3,3,5)
plt.imshow(a,cmap="gray")
plt.title('a - channel')
plt.axis('off')

plt.subplot(3,3,6)
plt.imshow(b,cmap="gray")
plt.title('b - channel')
plt.axis('off')

plt.subplot(3,3,7)
plt.imshow(Y2,cmap="gray")
plt.title('Y2 - channel')
plt.axis('off')

plt.subplot(3,3,8)
plt.imshow(Cr,cmap="gray")
plt.title('Cr - channel')
plt.axis('off')

plt.subplot(3,3,9)
plt.imshow(Cb,cmap="gray")
plt.title('Cb - channel')
plt.axis('off')

plt.show()
```

<h3>Inference</h3>
<ol>
    <li>Combination of <b>b</b> from CIELab color space, <b>V</b> from YUV color space and <b>Cr</b> from YCbCr color space works well for images with a clear distinction between background and salient part. </li>
    <li>Similarly, combination of <b>a</b> from CIELab color space, <b>U</b> from YUV color space and <b>Cb</b> from YCbCr color space works well for images with a blurred background. </li>
    </ol>
